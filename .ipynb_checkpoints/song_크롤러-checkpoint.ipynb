## 인기순 크롤러

```python
# 크롤러
### 페이지 접근
# 크롬드라이버 열기
driver = wd.Chrome() # 크롬드라이버 경로

# 드라이버가 해당 url 접속
url = 'https://www.melon.com/genre/song_list.htm?gnrCode=GN0800#params%5BgnrCode%5D=GN0800&params%5BdtlGnrCode%5D=&params%5BorderBy%5D=POP&params%5BsteadyYn%5D=N&po=pageObj&startIndex=1' # 포크 블루스 페이지 인기순 접근
driver.get(url)
time.sleep(2)

# 인기순 접근
# driver.find_element(By.XPATH,'//*[@id="frm"]/div/div/div/a[2]')

song_info = [] # 곡정보
img = []    # 앨범 사진
song = []   # 노래 이름
singer = [] # 가수
lyrics = [] # 가사
like = []   # 좋아요 수

for i in range(1, 50):
    
    driver.find_element(By.XPATH, f'//*[@id="frm"]/div/table/tbody/tr[{i}]/td[4]/div/a').click()
    
    
    img.append(driver.find_element(By.XPATH, '//*[@id="d_song_org"]/a/img').get_attribute("src"))
    song.append(driver.find_element(By.XPATH, '//*[@id="downloadfrm"]/div/div/div[2]/div[1]/div[1]').text)
    singer.append(driver.find_element(By.XPATH, '//*[@id="downloadfrm"]/div/div/div[2]/div[1]/div[2]/a').text)
    lyrics.append(driver.find_element(By.XPATH, '//*[@id="d_video_summary"]').text)
    like.append(driver.find_element(By.XPATH, '//*[@id="d_like_count"]').text)
    
    driver.get(url)
    time.sleep(3)
driver.quit()
```

```python
df = pd.DataFrame(zip(song, singer, lyrics, img, like))

df.columns = ['title', 'singer', 'lyrics', 'img', 'like']
```
